```python
import sys
import glob
import os
import re
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import datetime
%matplotlib inline
```

```python
def sum_throughput(trace):
    s = []
    for t in trace:
        trace = pd.read_table(t, header=None, sep=(","),names=("seconds", "rate"))
        if len(s) == 0: s = trace
        else: s += trace
    return s
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

s = sum_throughput(glob.glob("mantle-issdm*.csv"))
ax.plot(range(0, len(s['rate'])), s['rate'], 
        'r-', marker='o', linewidth=2.0, label="wave 1, seq 0")

s = sum_throughput(glob.glob("mantle-back*.csv"))
ax.plot(range(-1, len(s['rate']) - 1), s['rate'], 
        'b-', marker='o', linewidth=2.0, label="wave 1, seq 1")

ax.set_ylim(0, 2000)
ax.set_xlabel("seconds")
ax.set_ylabel("Operations per second")
lgd = ax.legend(loc='center', bbox_to_anchor=(1.5, 0.8))
fig = plt.figure()
ax = fig.add_subplot(111)

s = sum_throughput(glob.glob("mantle2-issdm*.csv"))
ax.plot(range(0, len(s['rate'])), s['rate'], 
        'r-', marker='x', linewidth=2.0, label="wave 2, seq 0")

s = sum_throughput(glob.glob("mantle2-background-issdm*.csv"))
ax.plot(range(0, len(s['rate'])), s['rate'], 
        'b-', marker='x', linewidth=2.0, label="wave 2, seq 1")

ax.set_ylim(0, 2000)
ax.set_xlabel("seconds")
ax.set_ylabel("Operations per second")
lgd = ax.legend(loc='center', bbox_to_anchor=(1.5, 0.8))
fig = plt.figure()
ax = fig.add_subplot(111)

s = sum_throughput(glob.glob("mantle3-issdm*.csv"))
ax.plot(range(0, len(s['rate'])), s['rate'], 
        'r-', marker='^', linewidth=2.0, label="wave 3, seq 0")

s = sum_throughput(glob.glob("mantle3-background-issdm*.csv"))
ax.plot(range(0, len(s['rate'])), s['rate'], 
        'b-', marker='^', linewidth=2.0, label="wave 3, seq 1")

ax.set_ylim(0, 2000)
ax.set_xlabel("seconds")
ax.set_ylabel("Operations per second")
lgd = ax.legend(loc='center', bbox_to_anchor=(1.5, 0.8))

s = sum_throughput(glob.glob("mantle-switch-issdm*.csv"))
ax.plot(range(0, len(s['rate'])), s['rate'], 
        'r-', marker='v', linewidth=2.0, label="wave 3, seq 0")

s = sum_throughput(glob.glob("mantle-switch-background-issdm*.csv"))
ax.plot(range(0, len(s['rate'])), s['rate'], 
        'b-', marker='v', linewidth=2.0, label="wave 3, seq 1")

ax.set_ylim(0, 2000)
ax.set_xlabel("seconds")
ax.set_ylabel("Operations per second")
lgd = ax.legend(loc='center', bbox_to_anchor=(1.5, 0.8))
```
